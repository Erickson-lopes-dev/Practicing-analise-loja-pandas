```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
loja1 = pd.read_csv('loja_1')
# loja1['loja'] = [f'l{num}' for num in range(61) ]
loja1['loja'] = 'l1'
loja1
```

```python
loja2 = pd.read_csv('loja_2')
loja2['loja'] = 'l2'
loja2
```

```python
loja3 = pd.read_csv('loja_3')
loja3['loja'] = 'l3'
loja3
```

```python
# concatenei os tres dataframes com os dados dos aquivos
df = pd.concat([loja1, loja2, loja3])

# Indiquei uma coluna como index e exibi 50 resultados aleatorios
df = df.set_index('id_cliente')
df
```

```python
# Renomenado nome da coluna
df = df.rename(columns={'quatidade': 'produto'})
df
```
